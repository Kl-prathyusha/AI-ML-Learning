```python
import numpy as np
```

```python
# data type of numpy is nd array which is a multi-dimensional array object
# numpy arrays can be created using the np.array() function


a= np.array(5)
print(a)  # Output: 5
print(a.dtype)  # Output: int64
print(a.shape)  # Output: ()
```

1. a single value is known as a scalar in numpy [5] - has no dimension
2. A group of scalar values is known as a vector in numpy [5, 10, 15] - has 1 dimension min
3. A group of vectors is known as a matrix in numpy [[5, 10, 15], [20, 25, 30]] - has min 2 dimensions
4. A group of matrices is known as a tensor in numpy [[[5, 10, 15], [20, 25, 30]], [[35, 40, 45], [50, 55, 60]]] - has min 3 dimensions

```python
#finding more info about the numpy array
print("datatype of a is:",a.dtype)
print("shape of a is:",a.shape) # number of rows and columns in the array
print("number of dimensions in a is:",a.ndim) # number of dimensions in the array
print("size of a is:",a.size) # total number of elements in the array   
print("itemsize of a is:",a.itemsize) # size of each element in the array in bytes
print("nbytes of a is:",a.nbytes) # total number of bytes consumed by the array
```
